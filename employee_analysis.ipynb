```python
import pandas as pd
df=pd.read_csv("employee__sql.csv")

print(df.shape)
print(df.columns)
print(df.dtypes)
print(df.isnull())
#
```

# Department Analysis

# Har department me kitne employees hain

```python
dept_count=df['department'].value_counts()
print(dept_count)
```

# Average salary by department

```python
dept_count=df.groupby('department')['salary'].mean()
print(dept_count)
```

# Average performance score by department

```python
dept_per=df.groupby('department')['performance_score'].mean().sort_values(ascending=False)
print(dept_per)
```

## Observation
##### 1. Employee distribution is fairly balanced across all departments.

##### 2. The IT department has the highest average salary.

##### 3. IT and Finance departments show better average performance scores.

# Employee count by city

```python
emp_count=df['city'].value_counts().sort_values(ascending=False)
print(emp_count)
```

# Average salary by city

```python
avg_salary=df.groupby('city')['salary'].mean().sort_values(ascending=False)
print(avg_salary)
```

#### Delhi and Bangalore have a significant share of employees, indicating major workforce concentration in these locations.

# Highest salary

```python
# Employee(s) with Highest Salary
df[df['salary'] == df['salary'].max()]
```

# Lowest salary

```python
df[df['salary']==df['salary'].min()]
```

#  Top 5 highest-paid employees

```python
df.sort_values(by='salary', ascending=False).head(5)
```

#### The salary distribution ranges from entry-level to senior-level compensation, with top salaries primarily observed in the IT department.

#  Average performance score

```python
df['performance_score'].mean()
```

# Highest performers

```python
df[df['performance_score']==df['performance_score'].max()]
```

# Department-wise performance

```python
df.groupby('department')['performance_score'].value_counts().sort_values(ascending=False)
```

### Insight
#### Employees with high performance scores are more common in IT  departments.

# Average experience

```python
df['experience'].mean()
```

# Most experienced employees

```python
df[df['experience']==df['experience'].max()]
```

# Experience vs salary relationship

```python
df.groupby('experience')[('salary')].mean()
```

### A positive relationship is observed between employee experience and salary growth.

```python
import matplotlib.pyplot as plt
```

# Department-wise Average Salary 

```python
dept_salary=df.groupby('department')['salary'].mean().sort_values(ascending=False)
```

```python
plt.bar(dept_salary.index,dept_salary.values,label='Salary', edgecolor='black')
plt.title(' Department-wise Average Salary (Bar Chart) ')
plt.xlabel('Department')
plt.ylabel('Salary')
plt.legend()
plt.savefig("Dept_wise_Salary_bar.png")
plt.show()
```

# Employee Distribution by Department

```python
dept_count=df.groupby('department')['department'].value_counts().sort_values(ascending=False)

plt.pie(dept_count.values,labels=dept_count.index,autopct='%1.1f%%')
plt.title('Employee Distribution by Department')
plt.savefig("Emp_dist_by_dept.png")
plt.show()
```

# Salary Distribution

```python
plt.hist(df['salary'], bins=30, color='lightgreen', edgecolor='black', alpha=0.8)
plt.title("Salary Disribution")
plt.xlabel("Salary")
plt.ylabel("No. of employee")
plt.savefig('Salary_dist.png')
plt.show()
```

#  Experience vs Salary

```python
plt.scatter(df['experience'],df['salary'])
plt.xlabel('experience')
plt.ylabel('salary')
plt.colorbar()
plt.savefig('experience_salary.png')
plt.show()
```

# Department-wise Average Performance Score

```python
dept_avg_per=df.groupby('department')['performance_score'].mean().sort_values(ascending=False)
plt.plot(dept_avg_per.index,dept_avg_per.values,marker='o')
plt.title('Department-wise Average Performance Score')
plt.xlabel('Department')
plt.ylabel('Performance Score')
plt.savefig('dept_wise_avg_per_score.png')
plt.show()
```
